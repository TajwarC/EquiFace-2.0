```python
import matplotlib.pyplot as plt
import numpy as np
import yaml

def plot_chart(yaml_path, model_name="ArcFace (tflite)", dataset_name="SynthPar", percentage=None):
    # Load YAML file
    with open(yaml_path, 'r') as f:
        yaml_data = yaml.safe_load(f)

    # Organize the data
    data = {}
    for entry in yaml_data:
        # Extract group label from dataset (e.g., "synthpar\ST1" -> "ST1")
        group = entry['dataset'].split('\\')[-1].upper()
        if group not in data:
            data[group] = {}
        # Store FNR and FPR
        if entry['metric'] == 'FNR':
            data[group]['FNR'] = entry['value']
        elif entry['metric'] == 'FPR':
            data[group]['FPR'] = entry['value']

    groups = list(data.keys())
    fnr_values = [data[group].get("FNR", 0) * 100 for group in groups]
    fpr_values = [data[group].get("FPR", 0) * 100 for group in groups]

    x = np.arange(len(groups))
    width = 0.4

    fig, ax = plt.subplots(figsize=(12, 6))
    bars_fnr = ax.bar(x - width/2, fnr_values, width, label='FNR (%)', color='firebrick')
    bars_fpr = ax.bar(x + width/2, fpr_values, width, label='FPR (%)', color='deepskyblue')

    ax.set_xlabel('Synthetic Groups')
    ax.set_ylabel('Rate (%)')
    ax.set_title(f'{model_name} FPR and FNR Across Synthetic Groups ({dataset_name}), ({percentage if percentage else "threshold unknown"})')
    ax.set_xticks(x)
    ax.set_xticklabels(groups)
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Add value labels
    def add_labels(bars):
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.2f}%',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=9)

    add_labels(bars_fnr)
    add_labels(bars_fpr)

    plt.tight_layout()
    plt.show()
```

```python
yaml_path = 'results_synthpar/VGGFace_25.yaml'
model_name = 'VGG Face'
dataset_name = 'SynthPar'
percentage = '25%/0.5%'

plot_chart(yaml_path, model_name, dataset_name, percentage)
```
